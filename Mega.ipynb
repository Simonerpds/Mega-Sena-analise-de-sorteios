Análise dos Sorteios da Mega Sena

```python
import pandas as pd
df = pd.read_excel("Mega Sena/Mega-Sena.xlsx")
```

```python
#1. 6 Números mais frequentes
# Juntar todas as bolas em uma lista
numeros = pd.concat([df['Bola1'], df['Bola2'], df['Bola3'],
                     df['Bola4'], df['Bola5'], df['Bola6']])

# Contar frequência
frequencia = numeros.value_counts().sort_index()

print(frequencia)
```

```python
#6 mais frequentes
top_6 = frequencia.sort_values(ascending=False).head(6)

print(top_6)
```

```python
#2. Pares x Ímpares
pares = numeros[numeros % 2 == 0].count()
impares = numeros[numeros % 2 != 0].count()

print("Pares:", pares)
print("Ímpares:", impares)
```

```python
#3. Espaçamento entre
espacamentos = []

for i in range(1, len(df)):
    anterior = df.loc[i-1, ['Bola1','Bola2','Bola3','Bola4','Bola5','Bola6']].values
    atual = df.loc[i, ['Bola1','Bola2','Bola3','Bola4','Bola5','Bola6']].values

    dif = abs(atual - anterior)
    espacamentos.extend(dif)

# Frequência dos espaçamentos
esp_freq = pd.Series(espacamentos).value_counts().sort_index()

print(esp_freq)
```

```python
#4. Padrões de espaçamento repetidos
padroes = []

for i in range(1, len(df)):
    anterior = df.loc[i-1, ['Bola1','Bola2','Bola3','Bola4','Bola5','Bola6']].values
    atual = df.loc[i, ['Bola1','Bola2','Bola3','Bola4','Bola5','Bola6']].values

    padrao = tuple(abs(atual - anterior))
    padroes.append(padrao)

padroes_freq = pd.Series(padroes).value_counts()

print(padroes_freq.head())
```

```python
#5. Número que mais saiu por posição
for i in range(1, 7):
    print(f"\nBola {i}:")
    print(df[f'Bola{i}'].value_counts().head(1))
```

```python
#6. Ordenar do MENOS frequente para o MAIS frequente
# Juntar todas as bolas em uma única lista
numeros = pd.concat([
    df['Bola1'], df['Bola2'], df['Bola3'],
    df['Bola4'], df['Bola5'], df['Bola6']
])

# Contar quantas vezes cada número apareceu
frequencia = numeros.value_counts()

# Ordenar do MENOS frequente para o MAIS frequente
menos_frequentes = frequencia.sort_values()

print(menos_frequentes)
```

```python
#7. Ver os 10 menos Frequentes
print(menos_frequentes.head(10))
```

```python
#8. Mostrar com número e quantidade
for numero, qtd in menos_frequentes.head(10).items():
    print(f"Número {numero} apareceu {qtd} vezes")
```

```python
#9. Números que nunca foram sorteados
# Criar lista com todos os números possíveis (1 a 60)
todos = set(range(1, 61))

# Números que já saíram
numeros = pd.concat([
    df['Bola1'], df['Bola2'], df['Bola3'],
    df['Bola4'], df['Bola5'], df['Bola6']
])

sorteados = set(numeros)

# Diferença = números que nunca saíram
nunca_sorteados = todos - sorteados

print("Números que nunca foram sorteados:")
print(sorted(nunca_sorteados))
```

```python
#10. Média dos números sorteados em cada concurso
# Calcular média de cada linha (sorteio)
df['Media'] = df[['Bola1','Bola2','Bola3','Bola4','Bola5','Bola6']].mean(axis=1)

# Média geral
media_geral = df['Media'].mean()

print("Média geral dos números sorteados:", media_geral)
```
